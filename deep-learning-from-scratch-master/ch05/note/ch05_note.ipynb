## 損失関数と活性化関数

誤差逆伝播法では「出力層から逆伝播する値=予測値-正解値（誤差）」になるような損失関数と出力層の活性化関数の組み合わせを選ぶ必要がある。

以下がその代表的な組み合わせである。
| 問題種別 | 活性化関数         | 損失関数             | 
| -------- | ------------------ | -------------------- | 
| 分類問題 | ソフトマックス関数 | 交差エントロピー関数 | 
| 回帰問題 | 恒等関数           | 二乗和誤差関数       |
