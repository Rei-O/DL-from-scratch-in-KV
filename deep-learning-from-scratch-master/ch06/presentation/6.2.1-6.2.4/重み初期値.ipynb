```python
# 単純なニューラルネットワークを作成

import numpy as np

def sigmoid(u):
    return 1 / (1 + np.exp(-u))

def softmax(u):
    e = np.exp(u)
    return e / np.sum(e)

def ReLU(u):
    return np.maximam(0, u)

def tanh(u):
    return np.tanh(u)

def forward(x, W1, W2):
    params = {}
    params['W1'] = W1
    params['W2'] = W2
    params['u1'] = x.dot(W1)
    params['z1'] = sigmoid(params['u1'])
    params['u2'] = params['z1'].dot(W2)
    params['y']= softmax(params['u2'])
    return params


def back_propagation(params, d):
    delta2 = params['y'] - d
    grad_W2 = params['z1'].T.dot(delta2)

    sigmoid_dash = params['z1'] * (1 - params['z1']) # シグモイド関数の微分
    delta1 = delta2.dot(params['W2'].T) * sigmoid_dash
    grad_W1 = x.T.dot(delta1)

    params['W2'] -= learning_rate * grad_W2
    params['W1'] -= learning_rate * grad_W1
```

```python
# パラメータ、入力値などを設定

hidden = 5
output = 3
x = np.random.randn(1, 4)
d = np.array([[1, 0, 0]])
learning_rate = 0.5
```

### Z1の値の分布

```python
import numpy as np
import matplotlib.pyplot as plt

input_data = np.random.randn(1000, 100)  # 1000個のデータ
node_num = 100  # 各隠れ層のノード（ニューロン）の数
hidden_layer_size = 1  # 隠れ層が5層
activations = {}  # ここにアクティベーションの結果を格納する

x = input_data

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]

    # 初期値の値をいろいろ変えて実験しよう！
    w = np.random.randn(node_num, node_num) * 1
    # w = np.random.randn(node_num, node_num) * 0.01
    # w = np.random.randn(node_num, node_num) * np.sqrt(1.0 / node_num)
    # w = np.random.randn(node_num, node_num) * np.sqrt(2.0 / node_num)


    a = np.dot(x, w)


    # 活性化関数の種類も変えて実験しよう！
    z = sigmoid(a)
    # z = ReLU(a)
    # z = tanh(a)

    activations[i] = z

# ヒストグラムを描画
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    if i != 0: plt.yticks([], [])
    # plt.xlim(0.1, 1)
    # plt.ylim(0, 7000)
    plt.hist(a.flatten(), 30, range=(0,1))
plt.show()
```

## →シグモイド関数の微分値がほとんど0になることで、前の層に伝える数値が小さくなり学習が進みにくい
