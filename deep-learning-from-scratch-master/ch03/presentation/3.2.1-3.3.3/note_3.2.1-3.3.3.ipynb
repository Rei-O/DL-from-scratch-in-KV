### 恒等関数
* 出力値の意味づけ：入力信号をそのまま伝える

### ステップ関数
* 出力値の意味づけ：ニューロンを発火するかしないか？
* 入力がしきい値より大きい場合1、以下の場合0を出力する。

■ 定義<br>
![](fig/ステップ関数.png)

```python
# ステップ関数を可視化
import numpy as np
import matplotlib.pylab as plt


def step_function(x):
    return np.array(x > 0, dtype=np.int)

X = np.arange(-5.0, 5.0, 0.1)
Y = step_function(X)
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)  # 図で描画するy軸の範囲を指定
plt.show()
```

### シグモイド関数
* 出力値の意味づけ：ステップ関数を「1である確率」に変換したもの
  - ステップ関数の連続近似関数として用いられる
  - ロジスティック関数の特別な場合なのでロジスティック分布の累積分布関数として確率を与える
* 微分が簡単（シグモイドレイヤの実装でやるので省略）
* 多クラス分類に拡張できる（ソフトマックス関数の導入でやるので省略）

■ 定義<br>
![](fig/シグモイド関数.png)

```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))    

X = np.arange(-5.0, 5.0, 0.1)
Y = sigmoid(X)
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### ReLU関数
* ニューロンを発火させるとき信号の強さも知りたい


■ 定義<br>
![](fig/ReLU関数.png)

```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt


def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()
```
